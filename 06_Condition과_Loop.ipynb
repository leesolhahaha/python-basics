## 조건문(Condition)
- 특정한 조건을 만족하면 (True) 코드를 실행
- if, elif, else

  - if : 만약에 ~라면 ~해라
  - elif (else if의 약어) : 그게 아니고 ~ 라면 ~ 해라
  - else : 그게 아니면 ~ 해라

## 조건문의 구조

````python
if [condition]:
  code context
elif [condition]:
  code context
elif [condition]:
  code context
...
else:
  code context
````

- 조건 부분에 bool타입 데이터(True/False)가 온다.

## Code Context
- Context : 코드문맥
- 특정한 상황에 실행해야할 코드 뭉터기
- 파이썬에는 context를 indent(들여쓰기)로 구분
- indent는 공백 2칸이나 4칸으로 설정하는 것이 관례
- 처음에 indent를 2칸으로 지정했으면 계속 2칸으로 쓰고, 4칸으로 지정했으면 계속 4칸으로 쓰면 된다.

````python
a = 3
b = 2

if a > b:
  print("hi")
  print("bye")

print("done") # 3개가 다 나온다고 해서 다 코드콘텍스트라고 생각하면 안됨 done은 빠져나온 거임
````

elif, else는 if에서 조건이 맞지 않았을 때 연속해서 조ㅗ건검사를 하고 싶을 때 사용한다.

  만약에 지갑에 12000원 이상 있으면 택시를 타고, 그게 아니고 지갑에 8000~12000원 미만 있으면 버스를 타고, 그게 아니면 걸어 간다.

````python
money = 15000

if money >= 12000:
  print("택시타기")
elif money >= 8000 and money < 12000:
  print("버스타기")
else:
  print("걸어가기")
````

# 반복문(Loop)

##while

````python
while [codition]
  code context
````

````python
data = 3

#data가 0이 아니면 (condition - 조건) 반복해라
while data != 0:
  print("Hello", data)
  data -= 1

#조건검사를 먼저하는게 아니라 코드컨텍스트를 하고 하는 것

print("done")
````

````python
# 강사님이 하신 거

# 1. 몇명의 점수를 입력 받을 것인가?
# 2. 명 수가 절해지면 그 만큼 반복하면서 점수를 입력 받음
#  - 점수를 입력받으면, 그 점수 정보를 저장 -> 컬렉션을 활용
# 3. 다 일벽 되면 총점과 평균을 구함

count = int(input("몇 명의 점수를 입력 받으시겠습니까?"))

scores = []

for i in range(count):

  print("-" * 10)
  math_score = int(input("수학 점수 : "))
  sci_score = int(input("과학 점수 : "))

  score = {
      "math": math_score,
      "science": sci_score
  }

  scores.append(score)

# 일단 리스트 형식으로 만들어서 정리하는 거 리스트안에 딕셔너리 구조로 들어있는 형식

math_sum, science_sum = 0, 0

for score in scores:
  math_sum += score["math"]
  science_sum += score["science"]

print(math_sum / len(scores))
print(science_sum / len(scores))

# 이 코드는 리스트안에 딕셔너리
````

````python
count = int(input("몇 명의 점수를 입력 받으시겠습니까?"))

math_score_list = []
science_score_list = []

score_dict = {}

for i in range(count):
  print("-" * 10)
  math_score = int(input("수학 점수 : "))
  sci_score = int(input("과학 점수 : "))

  math_score_list.append(math_score)
  science_score_list.append(sci_score)

score_dict['math'] = math_score_list
score_dict['science'] = science_score_list

print(score_dict)

# 이 코드는 딕셔너리 안에 리스트 형식으로 만듦
````

# 'continue, breack'

- continue : 반복문에서 해당 반복을 중지하고 계속 이어서 반복 루틴을 수행
- breack : 반복문 자체를 중지

컨티뉴는 와일문은 조건 검사
포문은 꺼내는 것 부터

둘 다 반복문내에 조건문 안에서 사용되는 것

````python
for item in range(5):

  if item == 2:  # 2를 만나는 순간 컨티뉴가 되면서 다시 꺼내러 가게 됨
    # print("continue를만납니다.", item) # 2일 때만 true이기 때문에 여기서 item하면 2가 나옴
    continue

  print(item) # 2일 때만 true이기 때문에 이 아이템이 출력되는 거는 2를 제외한 숫자들일 때만 이게 출력되고 출력되면 다시 컨티뉴되어서 위로가서 반복한다.
````

````python
for item in range(5):

  if item == 2:
    break # 브레이크를 하기 때문에 2가되면 멈춘다.

  print(item)
````

````python
# 1. 구구단 ver.2 난이도 ★★☆☆☆
  # * 홀수 단은 `~~ 곱하기 ~~ 는 ~~`
  # * 짝수 단은 `~~ x ~~ = ~~`
  # * 3단이면 `3 곱하기 1는 3`
  # * 4단이면 `4 x 1 = 4`

for dan in range(2, 10):
  print(f"==={dan}단===")

  for times in range(1, 10):
    if dan % 2 == 0:
      print(f"{dan} x {times} = {dan * times}")
    else :
      print(f"{dan} 곱하기 {times} 는 {dan * times}")
````

````python
# 2. 구구단 ver.3 난이도 ★★★☆☆
#   * 가로 구구단 출력하기
#   * ```
#     2*1=1 3*1=3 4*1=4 .....
#     2*2=4 3*2=6 4*2=8 .....
#   ```
#   * `print`함수 사용시에 `end="\t"`를 이용하면 옆으로 감

for times in range(1, 10):
  print()

  for dan in range(2, 10):
    print(f"{dan} X {times} = {dan * times}", end='\t')
````

````python
# 3. 구구단 ver.4 난이도 ★★★★★
#   * 격자 구구단 출력하기
#     2단   3단   4단
#     ...   ...   ...
#     5단   6단   7단
#     ...   ...   ...
#     8단   9단

# 격자 구구단.
# 가로로 구구단이 배치 되어야하고
# 단이 3개씩 묶이면 다음 단은 아래로 내려가야함
# 단이 시작 될 때 마다 몇단인지 알려줘야함

# list로 만들어서 스텝사용하기
dans = range(2, 10)
times = range(1, 10)

h_num = 3 # 가로로 표시할 숫자의 개수

for idx in range(0, len(dans), h_num):
  # 한 줄에 표시할 단 가져오기
  now_dan = dans[idx : idx + h_num] # 1) 0, 1, 2

  for dan in now_dan:
    print(f"{dan} 단", end='\t\t')
  print()

  for time in times:
    for dan in now_dan:
      print(f"{dan} * {time} = {dan * time}", end='\t')
    print()
  print()
````

## 삼(3)항 연산자

- 파이썬이 아닌 다른 언어들은 삼항연산자(?)가  존재하지만, 파이썬에는 없음
- if, else만 사용해서 삼항연산을 수행

간단한 if, else 구문을 한 줄의 코드로 표현할 수 있는 방법


````python
result = None # None: 아무것도 없음을 의미

a = 1
b = 2

if a > b:
  result = "Hello"
else:
  result = "Bye"

result
# 간단한 조건을 사용할 때 사용하기 좋음
# 못 쓰는 경우는 조건문에 대해서 두줄이상의 컨텍스트가 펼쳐질 때는 못씀
````

* 삼항연산 사용 방법
```python
<True 일 때 사용할 값> if [condition] else <False 일 때 사용할 값>
```

````python
result = "Hello" if a > b else "Bye"
result
````

# 컴프리헨션

- list, set, dict를 간단하게 만들기 위한 문법
- tuple은 컴프리헨션으로 만들 수 없다.

반복문을 이용해서 컬렉션에 데이터를 추가하는 경우가 많다!
- 여러 전처리나 효과없이 아주 단순하게 만들어진 데이터를 컬렉션에 추가(append)하고 싶으면??

- 한 줄짜리 간단한 코드에

````python
# 1~5까지 각각의 정수에 제곱한 결과를 리스트에 담아놓기

result = []

for data in range(1, 6):
  result.append(data ** 2)

result
````

`list` 컴프리헨션
```python
[ <추가할 값에 대한 처리> for <값> in <collection> if <값에 대한 조건> ]

<값에 대한 조건>이 True일 때 collection에 있는 아이템을 리스트에 추가
if문은 있어도 되고 없어도 된다.
```
코딩해야할 순서 ! ) 컬렉션에서 값을 뽑아서 값에 대한 조건을 보고 추가할 값에 대한 처리

````python
result = [ data ** 2 for data in range(1, 6) ] # 앞쪽부터 하지말고 포문부터 작성하기!
result
````

````python
# 1~ 100까지 홀수인 정수에 대해서만 제곱한 후에 추가
result = [data ** 2 for data in range(1, 101) if data % 2 == 1]
print(result)
````

````python
# 1~100까지 정수에 대해서 홀수는 제곱, 짝수는 그래도 리스트에 추가하는 컴프리헨션
result = [data if data % 2 == 0 else data ** 2 for data in range(1, 101)] #앞에는 삼항연산자를 사용하기 왜냐면 컴프리헨션에서는 else를 사용할 수 없기 때문에 오로지 if만
print(result)
````

컴프리헨션은 일반 for문을 사용한 것보다 빠르다

````python
%%timeit

lst = []

for data in range(1, 100001):
  if data % 3 == 0:
    lst.append(data)

len(lst)
````

````python
%%timeit
lst = [data for data in range(1, 100001) if data % 3 == 0]
len(lst)
````

## set과 dict 컴프리헨션

````python
text = 'aaabbbccc'
set(text)
````

````python
# set 컴프리헨션
result = {f"{letter}입니다."for letter in text}
result
````

````python
# dict 컴프리헨션
key_list = ["아침", "점심", "저녁"]
value_list = ["바나나", "구내식당", "카레"]

result = {meal : food for meal, food in zip(key_list, value_list)}
result
````

딕셔너리 컴프리헨션 실습
1. 단어의 글자가 몇번 등장했는지에 대한 딕셔너리 만들기
  - "letter" --> {'l' : 1, 'e' : 2, 't' : 2, 'r' : 1}
2. 특정 문장에 단어가 몇 번 등장했는지에 대한 딕셔너리 만들기
  - "Hello hi bye hello" --> {'hello' : 2, 'hi' : 1, 'bye' : 1}

````python
"aaabbbccccdddddd".count("dd")
````

````python
# 단어의 글자가 몇번 등장했는지에 대한 딕셔너리 만들기
text = 'letter'

result = { char : text.count(char) for char in set(text) } # set을 씌워주면 더 편리하게 할 수 있음
result
````

````python
text = "Hello hi bye hello"

result = { term : text.count(term) for term in set(text.split())} # set을 사용하면 중복된 결과를 거를 수 있음
result
````
